```python
real_data = pd.read_csv('/content/sample_stock_data.csv')
```

```python
real_data = real_data.rename(columns={
    'Open Price':'open_price',
    'Close Price':'close_price',
    'High Price':'high_price',
    'Low Price':'low_price',
    'Volume':'volume',
    'Verdict':'verdict'
})
```

```python
real_data
```

```python
from sdv.metadata import SingleTableMetadata
```

```python
metadata_dict = {
            "columns":{
                "open_price":{
                    "sdtype": "numerical",
                    "computer_representation": "Float"
                    },
                "close_price":{
                    "sdtype": "numerical",
                    "computer_representation": "Float"
                },
                "high_price":{
                    "sdtype": "numerical",
                    "computer_representation": "Float"
                },
                "low_price":{
                    "sdtype": "numerical",
                    "computer_representation": "Float"
                },
                "volume":{
                    "sdtype": "numerical",
                    "computer_representation": "Float"
                },
                "verdict":{
                    "sdtype":"categorical",
                }

            }
}
```

```python
metadata = SingleTableMetadata.load_from_dict(metadata_dict)
```

```python
from sdv.single_table import GaussianCopulaSynthesizer

synthesizer = GaussianCopulaSynthesizer(metadata)
```

```python
synthesizer.fit(
    data=real_data
)
```

```python
synthetic_data = synthesizer.sample(
    num_rows=5000
)

synthetic_data.head()
```

```python
synthetic_data.to_csv("dummy.csv",index=False)
```
